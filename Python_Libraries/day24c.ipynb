```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\sande\OneDrive\Documents\data.csv")
df
```

```python
plt.rcParams["figure.figsize"]=6,2 # Set figure size for better visibility
import warnings
warnings.filterwarnings("ignore")# ignore warnings to keep output clean
```

```python
vis=sns.distplot(df["InternetUsers"]) # Distribution plot - Univariate
```

```python
vis1=sns.displot(df["InternetUsers"])
```

```python
vis2=sns.displot(df["InternetUsers"],kde=True) # KDE Kernel Distribution curve
plt.show()
```

```python
vis2=sns.displot(df["InternetUsers"],kind="ecdf")
plt.show()
```

```python
vis2=sns.displot(data=df,x="InternetUsers",hue="IncomeGroup",kind="ecdf") # Hue only works when there is full data
```

```python
vis3=sns.distplot(df["BirthRate"],bins=18)
```

```python
vis4=sns.boxplot(x="IncomeGroup",y="InternetUsers",data=df) #Bivariate
```

```python
vis5=sns.lmplot(data=df,x="InternetUsers",y="BirthRate",fit_reg=False,hue="IncomeGroup")
```

```python
vis5=sns.lmplot(data=df,x="InternetUsers",y="BirthRate",hue="IncomeGroup") 
```
